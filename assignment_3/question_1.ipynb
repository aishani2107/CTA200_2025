```python
from iterate_complex import iterate_complex
import numpy as np
import cmath
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
# create data space
x = np.linspace(-2, 2, 500)
y = np.linspace(-2, 2, 500)
max_iter = 100
diverge_map1 = np.zeros((len(x), len(y)))
diverge_map2 = np.zeros((len(x), len(y)))

#fill in divergence maps
for i in range(len(x)):
    for j in range(len(y)):
        iter_count = iterate_complex(x[i], y[j], max_iter)
        #divergence map with iteration number when z diverges
        diverge_map2[j, i] = iter_count
        if iter_count < max_iter:
            #divergence map with bounded/diverging information
            diverge_map1[j, i] = 1

#plot diverge_map1
plt.figure(figsize=(5, 5), dpi=120)
plt.contourf(x, y, diverge_map1, cmap='magma', levels=1)
plt.xlabel("real (x)")
plt.ylabel("imaginary (y)")
plt.show()

#plot diverge_map2
plt.figure(dpi=120)
plt.contourf(x, y, diverge_map2, cmap='magma', levels=4)
plt.xlabel("real (x)")
plt.ylabel("imaginary (y)")
plt.colorbar(label="diverging iteration")
plt.show()
```
